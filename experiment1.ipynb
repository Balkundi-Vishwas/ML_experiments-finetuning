```python
import pandas as pd

date = [
        '2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', 
        '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'
    ]
car_type = ['SUV', 'Sedan', 'null', 'Truck', 'SUV', 'Sedan', 'Truck', 'null', 'SUV', 'Truck']
Speed = [60, 55, 58, 'null', 70, 'null', 50, 65, 68, 72]
Distance = [5, 7, 'null', 10, 8, 'null', 4, 9, 6, 11]
df = pd.DataFrame({
    "date":date,
    "car_type":car_type,
    "speed":Speed,
    "Distance":Distance,
})
```

```python
df["car_type"][0]
```

```python
lid=len(df['car_type'])
```

```python
for i in range(len(df['car_type'])):
    # print(df["car_type"][i])
    if df['car_type'][i]=="null":
        df['car_type'][i]=df['car_type'][lid-1]
```

```python
for i in range(len(df['speed'])):
    # print(df["car_type"][i])
    if df['speed'][i]=="null":
        df['speed'][i]=df['speed'][lid-1]
```

```python
for i in range(len(df['Distance'])):
    # print(df["car_type"][i])
    if df['Distance'][i]=="null":
        df['Distance'][i]=df['Distance'][lid-1]
```

```python
df['speed']
```

```python
for i in range(len(df['speed'])):
    if df['speed'][i]>60:
        df['target'][i]= 1
    else:
        df['target'][i]=0
```
